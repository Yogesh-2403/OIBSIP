```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C://Users//Yogesh Tewari//Downloads//advertising.csv")
```

```python
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
r2_score = reg.score(X_test, y_test)
```

```python
print(f'R-squared score: {r2_score:.2f}')
```

```python
plt.plot(y_test, color='blue')
plt.plot(y_pred, color='red')
plt.title('Actual vs Predicted Sales')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend(['Actual Sales', 'Predicted Sales'])
plt.show()
```

```python
plt.scatter(df['TV'], df['Sales'], color='red')
plt.scatter(df['Radio'], df['Sales'], color='blue')
plt.scatter(df['Newspaper'], df['Sales'], color='green')
plt.title('Advertising Spend vs Sales Revenue')
plt.xlabel('Advertising Spend')
plt.ylabel('Sales Revenue')
plt.legend(['TV', 'Radio', 'Newspaper'])
plt.show()
```

```python
corr = df[['TV', 'Radio', 'Newspaper', 'Sales']].corr()
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.title('Correlation Matrix')
plt.show()
```
